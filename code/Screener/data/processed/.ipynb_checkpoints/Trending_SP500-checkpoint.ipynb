```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
data = pd.read_csv('C:\\Users\\User\\Documents\\wiki\\wiki\\dev\\python\\Python-Stock\\code\\Screener\\data\\processed\\2022-07-26_future_commodities_nSP500_pvt_w_Fred.csv')
```

```python
dates = [pd.to_datetime(d).value for d in data.iloc[:,0]]
dates = np.array(dates)-np.array(dates[0])/10**11
```

```python
int_date_data = pd.concat([pd.DataFrame(dates),pd.DataFrame(data[['^GSPC']].values)],axis=1)
int_date_data.columns = ['Date','SP500']
```

```python
def trend_tau(set_):
    [x1, x2] = set_
    tau, p_value = stats.kendalltau(x1,x2)
    return([tau,p_value])
```

```python
def split_sequences(sequences, n_steps_in):

    X = list()
    for i in range(len(sequences)):
        # find the end of this pattern
        end_ix = i + n_steps_in
        # check if we are beyond the dataset
        # gather input and output parts of the pattern
        seq_x = sequences[i:end_ix, :]
        X.append(seq_x)

    return np.array(X)
```

```python
def trenddetector(set_, order=1):
    [list_of_index, array_of_data] = set_
    
    result = np.polyfit(list_of_index, list(array_of_data), order)
    slope = result[-2]
    return float(slope)
```

```python
n = 50
```

```python
windows = split_sequences(int_date_data.values,n)
windows = pd.DataFrame(windows).head(-(n-1)).values.ravel()
```
